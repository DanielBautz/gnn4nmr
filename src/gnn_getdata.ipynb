# Convert xyz to sdf

```python
!apt-get install -y openbabel
```

```python
import os
from google.colab import drive

# Step 1: Mount Google Drive if the files are on Google Drive
drive.mount('/content/drive')

# Set your base folder path here (Google Drive folder where your XYZ files are stored)
base_folder_path = '/content/drive/My Drive/Uni/Thesis/orca_xyz_formate'

# Set a new output folder for storing the converted SDF files
output_folder_path = '/content/drive/My Drive/Uni/Thesis/converted_sdf_files'

# Step 2: Create the output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)

# Step 3: Find all XYZ files in the directory and its subdirectories
xyz_files = []
for root, dirs, files in os.walk(base_folder_path):
    for file in files:
        if file.endswith(".xyz"):
            xyz_files.append(os.path.join(root, file))

print(f"Found {len(xyz_files)} XYZ files for conversion.")

# Step 4: Convert each XYZ file to SDF format and save in the output folder
for xyz_file in xyz_files:
    # Extract the base filename (without folder path)
    base_filename = os.path.basename(xyz_file).replace('.xyz', '')

    # Create the output SDF file path in the new output folder
    sdf_file = os.path.join(output_folder_path, base_filename + '.sdf')

    # Run the Open Babel command to convert the file
    !obabel "{xyz_file}" -O "{sdf_file}"

print("Conversion completed for all XYZ files. The SDF files are saved in:", output_folder_path)
```

```python
import py3Dmol
from rdkit import Chem
from rdkit.Chem import AllChem
import os

# Schleife durch alle SDF-Dateien mit dem Namen xxx_00.sdf
for i in range(1, 42):
    print(f"Molek√ºl: {i}")
    dateiname = f"/content/drive/MyDrive/Uni/Thesis/converted_sdf_files/{i:03}_00.sdf"
    if os.path.exists(dateiname):
        # Lade Molek√ºle aus der SDF-Datei
        supplier = Chem.SDMolSupplier(dateiname)
        molekuele = [mol for mol in supplier if mol is not None]

        # Erzeuge 3D-Koordinaten und zeige das erste Molek√ºl an
        if molekuele:
            mol = molekuele[0]
            # F√ºge Wasserstoffatome hinzu
            mol = Chem.AddHs(mol)
            # Embedding f√ºr die 3D-Koordinaten
            AllChem.EmbedMolecule(mol)

            # Hole die 3D-Koordinaten als PDB-String
            pdb_block = Chem.MolToPDBBlock(mol)

            # Verwende py3Dmol zur Darstellung
            viewer = py3Dmol.view(width=400, height=400)
            viewer.addModel(pdb_block, 'pdb')
            viewer.setStyle({'stick': {}})  # Stick-Darstellung
            viewer.addStyle({'atom': 'H'}, {'sphere': {'radius': 0.3}})  # F√ºgt Wasserstoffatome als Kugeln hinzu
            viewer.zoomTo()
            viewer.show()
        else:
            print(f"Keine Molek√ºle in der Datei {dateiname} gefunden.")
    else:
        print(f"Datei {dateiname} existiert nicht.")
```

```python
from rdkit import Chem
import networkx as nx
import matplotlib.pyplot as plt
from rdkit.Chem import rdmolops

# Funktion, um einen RDKit Molek√ºl-Graphen in einen NetworkX Graphen zu konvertieren
def mol_to_nx(mol):
    G = nx.Graph()

    # Atome als Knoten hinzuf√ºgen
    for atom in mol.GetAtoms():
        G.add_node(atom.GetIdx(), label=atom.GetSymbol())

    # Bindungen als Kanten hinzuf√ºgen
    for bond in mol.GetBonds():
        G.add_edge(bond.GetBeginAtomIdx(), bond.GetEndAtomIdx(), bond_type=bond.GetBondType())

    return G

# Molek√ºl aus einer SDF-Datei lesen
def read_sdf_file(file_path):
    suppl = Chem.SDMolSupplier(file_path)
    mols = [mol for mol in suppl if mol is not None]
    return mols

# Visualisierung des NetworkX Graphen
def draw_nx_graph(G):
    pos = nx.spring_layout(G)  # Layout f√ºr die Position der Knoten
    labels = nx.get_node_attributes(G, 'label')

    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, labels=labels, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
    plt.show()

# Hauptteil des Skripts
def main():
    sdf_file = "/content/drive/MyDrive/Uni/Thesis/converted_sdf_files/003_00.sdf"  # Pfad zur SDF-Datei
    mols = read_sdf_file(sdf_file)

    # Nur das erste Molek√ºl als Beispiel verwenden
    if len(mols) > 0:
        mol = mols[0]
        G = mol_to_nx(mol)
        draw_nx_graph(G)
    else:
        print("Keine g√ºltigen Molek√ºle in der SDF-Datei gefunden.")
```

```python
from rdkit import Chem
import networkx as nx
import matplotlib.pyplot as plt
from rdkit.Chem import rdmolops
import os

# Funktion, um einen RDKit Molek√ºl-Graphen in einen NetworkX Graphen zu konvertieren
def mol_to_nx(mol):
    G = nx.Graph()

    # Atome als Knoten hinzuf√ºgen
    for atom in mol.GetAtoms():
        G.add_node(atom.GetIdx(), label=atom.GetSymbol())

    # Bindungen als Kanten hinzuf√ºgen
    for bond in mol.GetBonds():
        G.add_edge(bond.GetBeginAtomIdx(), bond.GetEndAtomIdx(), bond_type=bond.GetBondType())

    return G

# Molek√ºl aus einer SDF-Datei lesen
def read_sdf_file(file_path):
    suppl = Chem.SDMolSupplier(file_path)
    mols = [mol for mol in suppl if mol is not None]
    return mols

# Visualisierung des NetworkX Graphen
def draw_nx_graph(G, title):
    pos = nx.spring_layout(G)  # Layout f√ºr die Position der Knoten
    labels = nx.get_node_attributes(G, 'label')

    plt.figure(figsize=(8, 6))
    plt.title(title)
    nx.draw(G, pos, with_labels=True, labels=labels, node_color='skyblue', node_size=500, font_size=10, font_weight='bold')
    plt.show()

# Hauptteil des Skripts
def main():
    for i in range(1, 42):
        sdf_file = f"/content/drive/MyDrive/Uni/Thesis/converted_sdf_files/{i:03d}_00.sdf"  # Pfad zur SDF-Datei im Format xxx_00.sdf
        if os.path.exists(sdf_file):
            mols = read_sdf_file(sdf_file)

            # Nur das erste Molek√ºl als Beispiel verwenden
            if len(mols) > 0:
                mol = mols[0]

                # Explizite Wasserstoffe hinzuf√ºgen
                mol = Chem.AddHs(mol)

                G = mol_to_nx(mol)
                draw_nx_graph(G, title=f"Molek√ºl aus Datei {i:03d}_00.sdf")
            else:
                print(f"Keine g√ºltigen Molek√ºle in der Datei {sdf_file} gefunden.")
        else:
            print(f"Datei {sdf_file} nicht gefunden.")
```
